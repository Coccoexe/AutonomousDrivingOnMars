```python
import os
import cv2 as cv
import numpy as np
import random
from tqdm.notebook import tqdm

#get the current working directory
cwd = os.getcwd()
#get the directory parent
FOLDER_MAIN = os.path.dirname(cwd)
#concatenate dataset\ai4mars-dataset-merged-0.1\msl\images\edr\train
FOLDER_DATASET = os.path.join(FOLDER_MAIN, 'dataset', 'ai4mars-dataset-merged-0.3', 'msl')
DIR_TRAIN_IMAGES = os.path.join(FOLDER_DATASET,'images', 'edr')
DIR_TRAIN_LABELS = os.path.join(FOLDER_DATASET,'labels', 'train')

DIR_TEST_IMAGES = os.path.join(FOLDER_DATASET,'images', 'edr')
DIR_TEST_LABELS = os.path.join(FOLDER_DATASET,'labels', 'test')

DIR_TEST_LABELS_MIN1 = os.path.join(DIR_TEST_LABELS,'masked-gold-min1-100agree')
DIR_TEST_LABELS_MIN2 = os.path.join(DIR_TEST_LABELS, 'masked-gold-min2-100agree')
DIR_TEST_LABELS_MIN3 = os.path.join(DIR_TEST_LABELS, 'masked-gold-min3-100agree')

print("done ...")
print("FOLDER_MAIN: ", DIR_TRAIN_IMAGES)

print(cv.__version__)
```

```python
#list all the files in the directory
files = os.listdir(DIR_TRAIN_IMAGES)

#delete last three letter of the file name
files = [f[:-4] for f in files]
print (files[1])
```

```python
n_3_images = 0
n_3_images_seen = 0

for f in files:
    #concatenate the file name with the directory
    img = cv.imread(os.path.join(DIR_TRAIN_IMAGES, f + '.jpg'))
    label = cv.imread(os.path.join(DIR_TRAIN_LABELS, f + '.png'))
    #chek if lable as at leat a pixel = 3
    if np.any(label == 3):
        #random number betwnn 0 and 10
        p = random.randint(0,9)
        n_3_images += 1
        if(p<1):
            #for each pixel in the label  with value 3, set the pixel to 255 0 0 (red)
            label[np.where((label == [3, 3, 3]).all(axis = 2))] = [0, 0, 255]
            #overlay the image with the label
            overlay = cv.addWeighted(img, 1, label, 0.3, 0)
            cv.imshow('overlay', overlay)
            cv.waitKey(0)
            n_3_images_seen += 1
            print("found a label with at least a pixel = 3 in ", f, "in", n_3_images)
    
    
```

```python

#print the percentage of the number of images with at least a pixel = 3

print("Number of images with big-rock: ", n_3_images, len(files))
print("Percentage of images with bigrock: ", (n_3_images/len(files))*100)
print("Number of images with big-rock seen: ", n_3_images_seen)
print("Percentage of images with real bigrock: ", (69/n_3_images_seen)*100)
```
