```python
import cv2 as cv
import os
import numpy as np

path_s5mars = os.path.join(os.path.dirname(os.path.join(os.getcwd())), 'dataset','S5Mars')
path_s5_images_train = os.path.join(path_s5mars, 'images', 'train')
path_s5_images_test = os.path.join(path_s5mars,'images','test')
path_s5_images = os.path.join(path_s5mars,'images')
path_s5_labels_train = os.path.join(path_s5mars, 'labels', 'train')
path_s5_labels_test = os.path.join(path_s5mars,'labels','test')
print(path_s5_labels_test)


def overlap(path_image,folder_lbl) :
    image = cv.imread(path_image)
    name = path_image.split('\\')[-1].split('.')[0]
    lbl_path = os.path.join(folder_lbl,name+'.png')

    lbl = cv.imread(lbl_path)
    #map color to class
    labels = np.array([[0,0,0],[1,1,1],[2,2,2],[3,3,3],[4,4,4],[5,5,5],[6,6,6],[7,7,7],[8,8,8],[9,9,9],[255,255,255]])
    colors = np.array([[255, 255, 0],[0,255,0],[0,255,255],[255,0,0],[255,0,255],[255,255,0],[255,255,255],[0,125,125],[0,125,125],[125,125,125],[125,125,0]])
    print(labels.shape)
    print(len(colors))
    for i in range(9):
        mask = np.all(lbl == labels[i], axis=-1)
        print(mask)
        lbl[mask] = colors[i]
    image = cv.addWeighted(image,1,lbl,0.5,0)
    cv.imshow('image',image)
    cv.waitKey(0)
    cv.destroyAllWindows()

```

```python
for file in os.listdir(path_s5_images_train):
    path_image = os.path.join(path_s5_images_train,file)

    overlap(path_image,path_s5_labels_train)
```
